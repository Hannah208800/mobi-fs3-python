```python
def meanfilter(img_in, size):
    img_mod=img_in.copy()
    result = numpy.zeros(img_in.shape)
    n = (size - 1) // 2
    for p in numpy.ndindex(img_in.shape):
        neighborhood_sum = 0
        if p[1] >= n and p[1] < img.shape[1]-n and p[0]>=n and p[0] < img.shape[0]-n:
            for q in numpy.ndindex(size, size):
                neighborhood_sum = neighborhood_sum + img_mod[p[0]-n+q[0],p[1]-n+q[1]]
        result[p]=neighborhood_sum/size**2
    return result

ima= img.copy()
ima=meanfilter(ima,5)
plt.imshow(ima)

```

## Non-linear filtering

**<a id="#">Task 3.2.1.</a>** Implement a median filter:

```python
import statistics
def medianfilter(img_in, size):
    img_mod=img_in.copy()
    result = numpy.zeros(img_in.shape)
    n = (size - 1) // 2
    for p in numpy.ndindex(img_in.shape):
        vektor=[]
        if p[1] >= n and p[1] < img.shape[1]-n and p[0]>=n and p[0] < img.shape[0]-n:
            for q in numpy.ndindex(size, size):
                vektor.append(img_mod[p[0]-n+q[0],p[1]-n+q[1]])
        result[p]=statistics.median(vektor)
    return result

imj= img.copy()
imj=medianfilter(ima,5)
plt.imshow(imj)
```
