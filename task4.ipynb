**<a id="#">Task 4.1.2.</a>** Implement horizontal and vertical 3&times;3 Prewitt filters:

```python
def prewitt_h(img_in):
    kernel = np.array([
        [1, 0, -1],
        [1, 0, -1],
        [1, 0, -1]
    ])

    result = np.zeros((img_in.shape[0]-2, img_in.shape[1]-2))

    for p in np.ndindex(result.shape):
        wert_neu = 0
        for k in np.ndindex((3,3)):
            wert_neu += kernel[k] * img_in[p[0]+k[0], p[1]+k[1]]
        result[p] = wert_neu

    return result


        
    
def prewitt_v(img_in):
    kernel = np.array([
        [1, 1, 1],
        [0, 0, 0],
        [-1, -1, -1]
    ])

    result = np.zeros((img_in.shape[0]-2, img_in.shape[1]-2))

    for p in np.ndindex(result.shape):
        wert_neu = 0
        for k in np.ndindex((3,3)):
            wert_neu += kernel[k] * img_in[p[0]+k[0], p[1]+k[1]]
        result[p] = wert_neu

    return result
```
