## Writing re-usable code (bonus)

**<a id="#">Task 2.3.4.</a>** Finish the implementation of the function `clip_image` below, by replacing the `...` line with your code:

```python
def clip_image(img_in, T1, T2):
    img_out = img_in.copy()
    img_prc=img_out.flatten()

    for i in range(len(img_prc)):
        if img_prc[i]<T1:
            img_prc[i]=T1
        if img_prc[i]>T2:
            img_prc[i]=T2

    img_out = img_prc.reshape(img_in.shape)
    return img_out

plt.imshow(clip_image(img,0.31,0.59),'gray')
plt.colorbar()
```
